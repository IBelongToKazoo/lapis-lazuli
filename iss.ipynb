Coordinates and Timestamp of the International Space Station (ISS)

```python
import requests
import time

def ISS_location():
    ISS_Api = requests.get('https://api.wheretheiss.at/v1/satellites/25544')

    #extract JSON data to get the coordinates (longitude, latitude) and the timestamp
    ISS_data = ISS_Api.json() 

    #set up the variables of the longitude, latitude, and timestamp
    latitude = float("{:.2f}".format(ISS_data["latitude"]))
    longitude = float("{:.2f}".format(ISS_data["longitude"]))
    timestamp = int(ISS_data["timestamp"])

    #output
    return latitude, longitude, timestamp

#100 data points
datapoints = []

for _ in range(100):
    latitude, longitude, timestamp = ISS_location()
    datapoints.append((latitude, longitude, timestamp))
    print(f"latitude: {latitude}, longitude: {longitude}, timestamp: {timestamp}")
    time.sleep(10)
```
